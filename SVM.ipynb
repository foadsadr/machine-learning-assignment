# svm کتابخانه های مورد نیاز برای  

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn import svm, datasets
```

# اضافه کردن دیتاست گل زنبق با استفاده فقط از دو فیچر

```python
iris = datasets.load_iris()
X = iris.data[:, :2]  
                     
y = iris.target

h = .02  
```

```python
df = pd.read_csv('iris.csv', header=None)
df.head()
```

#  و فیت کردن دیتاست svmایجاد چهار تابع مورد نیاز در

```python
C = 1.0  
svc = svm.SVC(kernel='linear', C=C).fit(X, y)
rbf_svc = svm.SVC(kernel='rbf', gamma=0.7, C=C).fit(X, y)
poly_svc = svm.SVC(kernel='poly', degree=3, C=C).fit(X, y)
lin_svc = svm.LinearSVC(C=C).fit(X, y)
```

# تابع برای رسم نمودار min/ maxایجاد کردن گره مشخص کردن  

```python
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
```

# انتخاب نام توابع برای رسم نمودار

```python
titles = ['SVC with linear kernel',
          'LinearSVC (linear kernel)',
          'SVC with RBF kernel',
          'SVC with polynomial (degree 3) kernel']
```

#  و رسم هر چهار مدل برای مقایسهdecision boundary رسم 

```python
for i, clf in enumerate((svc, lin_svc, rbf_svc, poly_svc)):
   
    plt.subplot(2, 2, i + 1)
    plt.subplots_adjust(wspace=0.4, hspace=0.4)

    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

  
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)
     
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm)
    plt.xlabel('Sepal length')
    plt.ylabel('Sepal width')
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.xticks(())
    plt.yticks(())
    plt.title(titles[i])

plt.show()
```
